```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("D:\practice_data_set\canada_per_capita_income.csv")
df.head()
```

```python
plt.scatter(df['year'],df['per capita income (US$)'],c='r',marker='*')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(df[['year']],df['per capita income (US$)'])
```

```python
predict = model.predict(df[['year']])
predict
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(df['year'],df['per capita income (US$)'],c='r',marker='*')
plt.plot(df['year'],predict,c='g')
plt.xlabel('Year')
plt.ylabel('Per Capita Income(US$)')
plt.show()
```
